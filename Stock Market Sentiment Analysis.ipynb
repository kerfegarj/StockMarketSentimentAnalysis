```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\Admin\Desktop\Kerfegar\Analytics&ML\Projects\Stock Market Sentiment Analysis\newsheadlines.csv', encoding="ISO-8859-1")
df.head(5)
```

```python
sns.countplot(df['Label'])
```

```python
x=df.drop(["Date","Label"],axis=1)
y=df["Label"]
x.head()
```

```python
x.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
```

```python
for i in x:
    x[i]=x[i].str.lower()
```

```python
headlines=[]
for i in range (0,len(x)):
    headlines.append(' '.join(str(x) for x in x.iloc[i,0:25]))
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
countvector=CountVectorizer(ngram_range=(2,2))
x=countvector.fit_transform(headlines)
```

```python
np.random.seed(50)
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
```

```python
from sklearn import svm
model=svm.SVC()
model.fit(xtrain,ytrain)
```

```python
ypreds=model.predict(x)
model.score(xtest,ytest)
```

```python
compare=pd.DataFrame({
    "Actual":df["Label"],
    "Predicted":ypreds
})
```
